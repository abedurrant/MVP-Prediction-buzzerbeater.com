# Combine Tables to create training data

```python
#Add %, PD, and MVP Points to league_leaders from season_standings_big_8, season_standings_great_8, mvp_votes
season_standings_big_8 = season_standings_big_8[["Big 8.1", "%", "PD", "season"]]
season_standings_great_8 = season_standings_great_8[["Great 8.1", "%", "PD", "season"]]
mvp_votes = mvp_votes[[0,1,7,"season"]]
```

```python
import re
mvp_votes['fixed name'] = mvp_votes[0].apply(lambda x: re.sub(' +', ' ', str(x)))
```

```python
#Combining based on the first letters of team name (as sometimes it is cut off in tables)
league_leaders[1] = league_leaders[1].str[:19]
season_standings_great_8['Great 8.1'] = season_standings_great_8['Great 8.1'].str[:19]
season_standings_big_8['Big 8.1'] = season_standings_big_8['Big 8.1'].str[:19]
new_df = pd.merge(league_leaders, season_standings_great_8,  how='left', left_on=[1,'season'], right_on = ['Great 8.1','season'])
new_new_df = pd.merge(new_df, season_standings_big_8,  how='left', left_on=[1,'season'], right_on = ['Big 8.1','season'])
finalish_df = pd.merge(new_new_df, mvp_votes,  how='left', left_on=[0,'season'], right_on = ['fixed name','season'])
```

```python
finalish_df.head(5)
```

```python
finalisher_df = finalish_df[['key_0',            2,            3,
                  4,            5,            6,        '7_x',            8,
                  9,           10,           11,           12,           13,
                 14,           15,        '%_x',
             'PD_x',         '%_y',       'PD_y',       '7_y',]]
```

```python
finalisher_df.rename(columns={'key_0': "Player",            2:"GP",            3:"MIN",
                  4:"FG",            5:"3FG",            6:"FT",        '7_x':"OR",            8:"TR",
                  9:"AST",           10:"TO",           11:"STL",           12:"BLK",           13:"PF",
                 14:"PTS",           15:"Rating",        '%_x':"Win %",
             'PD_x':"PD",         '%_y':"Win % 2",       'PD_y':"PD 2",       '7_y':"MVP Votes"}, inplace=True)
```

```python
closer_df = finalisher_df[finalisher_df['Player']!="Player"]
```

```python
#Combine columns into one and rename
closer_df['Win %'].fillna(closer_df['Win % 2'], inplace=True)
closer_df['PD'].fillna(closer_df['PD 2'], inplace=True)
final_df = closer_df[['Player', 'GP', 'MIN', 'FG', '3FG', 'FT', 'OR', 'TR', 'AST', 'TO',
       'STL', 'BLK', 'PF', 'PTS', 'Rating', 'Win %', 'PD', 'MVP Votes']]
```

```python
#Fill NA's and scale GP
final_df['MVP Votes'].fillna(0, inplace = True)
final_df['GP'] = final_df['GP'].astype(float).div(22)
```

```python
final_df.to_csv("traindataBBMVP.csv", index = False)
```
